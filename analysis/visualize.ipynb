```python
from visualize import *
```

```python
print("\n=== Coletando dados do Neo4j ===")
counts, rels, avg_deg = fetch_graph_stats()
degrees, num_nodes     = fetch_degree_distribution()
component_sizes        = fetch_wcc()
scc_sizes              = fetch_scc()

n_articles = counts["Article"]
n_cites    = rels["CITES"]

print("\n=== Tamanho do grafo (foco em Article + CITES) ===")
print(f"  Vértices (Article):  {n_articles:,}")
print(f"  Arestas  (CITES):    {n_cites:,}")
print(f"  Grau médio:          {avg_deg:.4f}")
print("\n  Outros nós:")
for label in ("Author", "Institution", "Venue", "Subfield"):
    print(f"    {label}: {counts[label]:,}")
print("\n  Outras arestas:")
for rel in ("AUTHORED_BY", "PUBLISHED_IN", "HAS_SUBFIELD", "AFFILIATED_WITH"):
    print(f"    {rel}: {rels[rel]:,}")

print("\n=== Componentes fracamente conexas ===")
print(f"  Total de componentes: {len(component_sizes)}")
print(f"  Maior componente:     {component_sizes[0]:,} vértices")
if len(component_sizes) > 1:
    print(f"  Menor componente:     {component_sizes[-1]} vértice(s)")
    print(f"  Componentes unitárias (k=1): {sum(1 for c in component_sizes if c == 1)}")

print("\n=== Componentes fortemente conexas ===")
print(f"  Total de componentes: {len(scc_sizes)}")
print(f"  Maior componente:     {scc_sizes[0]:,} vértices")
if len(scc_sizes) > 1:
    print(f"  Menor componente:     {scc_sizes[-1]} vértice(s)")
    print(f"  Triviais (k=1):       {sum(1 for c in scc_sizes if c == 1)}")
```

```python
plot_degree_distribution(degrees, num_nodes)
plot_component_sizes(component_sizes)
plot_scc_sizes(scc_sizes)
```
